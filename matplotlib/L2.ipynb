```python
gas = pd.read_csv("gas_prices.csv")
gas.set_index("Year", inplace=True)
gas
```

```python
x = list(gas.index)
print(x)
for country in gas:
    print(country, type(country))
```

```python
plt.title("Gas Prices")
plt.plot(gas.index, gas.USA, label = "USA", color = "black", marker = ".", markeredgecolor = "black")
plt.plot(gas.index, gas["Canada"], label = "Canada", color = "y", marker = ".", markeredgecolor = "y")
plt.plot(gas.index, gas["France"], label = "France", color = "b", marker = ".", markeredgecolor = "b")
plt.plot(gas.index, gas.Australia, label = "Australia", color = "r", marker = ".", markeredgecolor = "r")
plt.plot(gas.index, gas.Germany, label = "Germany", color = "g", marker = ".", markeredgecolor = "g")
plt.plot(gas.index, gas.Japan, label = "Japan", color = "pink", marker = ".", markeredgecolor = "pink")
plt.plot(gas.index, gas.Mexico, label = "Mexico", color = "purple", marker = ".", markeredgecolor = "purple")
plt.plot(gas.index, gas["South Korea"], label = "South Korea", color = "orange", marker = ".", markeredgecolor = "orange")
plt.plot(gas.index, gas.UK, label = "UK-gas", color = "violet", marker = ".", markeredgecolor = "violet")
plt.xlabel("Year")
plt.ylabel("Values in $")
plt.legend()
plt.show()
```

```python
plt.title("Gas Prices")
plt.bar(gas.index, gas.USA, label = "USA", color = "black", width=.4)
plt.bar(gas.index, gas.UK, label = "UK", width=.2)
plt.xlabel("Year")
plt.ylabel("Values in $")
plt.legend()
plt.show()
```

```python
plt.title("Gas Prices")
for country in gas:
    plt.plot(gas.index, gas[country], marker = ".", label = country)
plt.xlabel("Year")
plt.ylabel("Values in $")
plt.legend()
plt.savefig("gass-Price.png")
plt.show()
```

```python
# fifa.Weight = [int(x) if type(x) == str else x for x in fifa.Weight]
fifa.replace(to_replace={"Weight": 'lbs'}, value='', regex=True, inplace=True)
fifa["Weight"].dropna(inplace=True)
fifa["Weight"] = fifa["Weight"].astype(int)
fifa.Weight
```

```python
light = fifa.Weight.__getitem__(fifa.Weight < 125).count()
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.Weight.__getitem__(fifa.Weight >= 150 & fifa.Weight < 175).count()
medium_high = fifa.Weight.__getitem__(fifa.Weight >= 175 & fifa.Weight < 200).count()
high = fifa.loc(fifa.Weight >= 200).count()[0]
```
