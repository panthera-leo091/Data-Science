***Bar Chart***

```python
labels = ["A", "B", "C", "D", "E"]
values = [3, 5, 2, 9, 7]
# plt.bar(values, labels)
plt.title("Hamra Bar Graph")
bars = plt.bar(labels, values, color = "pink", label = "bar-graph")
bars[0].set_hatch("/")
bars[1].set_hatch("+")
bars[2].set_hatch("O")
bars[3].set_hatch(".")
bars[4].set_hatch("*")
plt.xlabel("X me ham h")
plt.ylabel("Y me ham h")
plt.show()
```
